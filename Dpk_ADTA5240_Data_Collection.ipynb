### We need to pass as an argument of Ticker i.e. the company’s ticker.

<strong>Note:</strong> A stock symbol or a ticker is a unique series of letters assigned to a security for trading purposes. For example:

- For Amazon, it is “AMZN”
- For Meta, it is “META”
- For Google, it is “GOOGL”

```python
import yfinance as yahooFinance

# META financial information
get_META = yahooFinance.Ticker("META")

# whole python dictionary is printed here
#print(get_META.info)

get_META_info = get_META.info
for key, value in get_META_info.items():
    print(key,':', value)
```

```python
# META key metrics

# display Company Sector
print("Company Sector : ", get_META_info['sector'])

# display Price Earnings Ratio
print("Price Earnings Ratio : ", get_META_info['trailingPE'])

# display Company Beta
print("Company Beta : ", get_META_info['beta'])
```

```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

### Store all news text in a single string

```python

# IMPORT ALL THE REQUIRED LIBRARIES
from bs4 import BeautifulSoup as BS
import requests as req

url = "https://www.businesstoday.in/latest/economy"

webpage = req.get(url)
trav = BS(webpage.content, "html.parser")
news_text = ''
for link in trav.find_all('a'):
    if(str(type(link.string)) == "<class 'bs4.element.NavigableString'>" and len(link.string) > 5):
        news_text += link.string + '\n'

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(news_text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Changing optional word cloud arguments

Now, change some optional arguments of the word cloud like max_font_size, max_word, and background_color.



```python
# lower max_font_size, change the maximum number of word and lighten the background:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(news_text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

### We can use the process_text() and words_ methods to display the word count and relative counts from the text respectively.

Source: https://towardsdatascience.com/generate-meaningful-word-clouds-in-python-5b85f5668eeb

```python
# create a dictionary of word frequencies
text_dictionary = wordcloud.process_text(news_text.upper())
# sort the dictionary
word_freq={k: v for k, v in sorted(text_dictionary.items(),reverse=True, key=lambda item: item[1])}

# use words_ to print relative word frequencies
rel_freq=wordcloud.words_

# print top 5 results
print(list(word_freq.items())[:5])
print(list(rel_freq.items())[:5])

# print all results
word_pd = pd.DataFrame(word_freq.items())
word_pd.to_csv('news_words.csv')
word_pd
```

## Example 3 Scrape Top 100 Movie List From IMDB
Source: https://www.stratascratch.com/blog/4-data-collection-libraries-in-python-that-you-should-know/

There are 50 records in each page so collect them separately

```python
# First 50 movies
#import pandas as pd
import requests
from bs4 import BeautifulSoup

movie_names = []
ratings = []

# from 51 to 100
# url = "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt"

# URL of the IMDb's top 100 movies page
url = "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"
headers = {"User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:123.0) Gecko/20100101 Firefox/123.0"}

# Fetch the content of the web page
response = requests.get(url, headers=headers)


# Parse the HTML content of the page
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the <div> elements with the class "lister-item mode-advanced"
movies = soup.find_all('li', class_='ipc-metadata-list-summary-item')


# For each movie, extract the title and rating
for movie in movies:

    title = movie.h3.text
    title = ' '.join(title.split()[1:])
    movie_names.append(title)

    rating = movie.find("span", attrs={'aria-label': True}).text.split()[0]
    ratings.append(rating)


movie_dic = {"Movie Name" : movie_names, "Rating" : ratings}
df = pd.DataFrame(movie_dic)
df
```

### Scrape a given number of movies and ratings


```python
# Determine the number of movies by a variable
# Not all values can be used
import requests
from bs4 import BeautifulSoup

movie_names = []
ratings = []
how_many = 1000

# Set the URL to scrape first 50
first_url = "https://www.imdb.com/search/title/?groups=top_" + str(how_many)+"&sort=user_rating,desc"
headers = {"User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:123.0) Gecko/20100101 Firefox/123.0"}


page_no = 1
next_page = ""

while page_no < how_many:
    current_page = first_url + next_page
    print(current_page)
    page = requests.get(current_page, headers=headers)

    # Parse the HTML content of the page
    soup = BeautifulSoup(page.content, 'html.parser')

    #Find the movies with the class
    movies = soup.find_all('li', class_='ipc-metadata-list-summary-item')

    # For each movie, extract the title and rating
    for movie in movies:

        title = movie.h3.text
        title = ' '.join(title.split()[1:])
        movie_names.append(title)

        rating = movie.find("span", attrs={'aria-label': True}).text.split()[0]
        ratings.append(rating)

    page_no += 50
    next_page = "&start=" + str(page_no) + "&ref_=adv_nxt"


movie_dic = {"Movie Name" : movie_names, "Rating" : ratings}
df = pd.DataFrame(movie_dic)
df
```

```python
# Save the movies and ratings in a csv file
df.to_csv('movies_ratings_1000.csv')
```

```python
import requests
from bs4 import BeautifulSoup



#URL of the Metacritic movie list page (e.g., Top Movies, New Releases, etc.)
url = 'https://www.metacritic.com/browse/movies/score/metascore/all/filtered'
base_url = 'https://www.metacritic.com/browse/movie/?releaseYearMin=1910&releaseYearMax=2024&page='

num_pages = 5
page = 0
movie_names_metacritic = []
ratings_metacritic = []

# Send GET request to the page
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}


for page in range(num_pages):  # Including the first page

      paginated_url = f'{base_url}'+ str(page)

      # Send GET request to the paginated page
      response = requests.get(paginated_url, headers=headers)
      #print(response)

      # Ensure the request was successful
      if response.status_code == 200:
          # Parse the HTML content
          soup = BeautifulSoup(response.text, 'html.parser')
          # Find all divs with the class 'c-finderProductCard'
          finder_product_cards = soup.find_all('div', class_='c-finderProductCard')


          # Loop through each product card and extract the data-title attribute
          for card in finder_product_cards:
              # Find the div with the data-title attribute within each card
              title_div = card.find('div', attrs={'data-title': True})

              if title_div:
                  # Extract the value of the data-title attribute
                  movie_title = title_div['data-title']
                  movie_names_metacritic.append(movie_title)

                  score_span = card.find('span', attrs={'data-v-e408cafe': True})

                  # Get the text from the span
                  if score_span:
                      score = score_span.text.strip()
                      ratings_metacritic.append(score)

                  else:
                      print('Metascore not found.')

      else:
          print(f"Failed to retrieve data. Status code: {response.status_code}")

metacritic_dic = {"Movie Name" : movie_names_metacritic, "Rating" : ratings_metacritic}
df = pd.DataFrame(metacritic_dic)
df
```
